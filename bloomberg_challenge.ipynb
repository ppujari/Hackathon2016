```python
import pandas as pd
import json
```

```python
# Load JSON directly (since it's a list of dictionaries)
with open('./data/LLMLawBriefArgumentLinking/stanford_hackathon_brief_pairs.json') as f:
    data = json.load(f)
```

```python
# Convert to DataFrame
df = pd.json_normalize(data)
```

```python
df.head()
```

```python
# Explode the moving brief arguments
df_exploded = df.explode('moving_brief.brief_arguments')

# Expand the argument dictionaries
df_exploded = df_exploded.join(
    pd.json_normalize(df_exploded['moving_brief.brief_arguments'])
)

# Drop the original nested column
df_exploded = df_exploded.drop(columns=['moving_brief.brief_arguments'])
```
